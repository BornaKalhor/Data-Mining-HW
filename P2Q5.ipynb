```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyfpgrowth
```

```python
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/divar_dataset/divar_posts_dataset.csv')
```

```python
df.head(3)
```

```python
df=df[['platform','cat3']]
df
```

```python
gpbycat = df.groupby(by="cat3")
transactions = []
gpbycat.filter(lambda x: (transactions.append(x['platform'].to_list())) if x.shape[0] >= 2 else True))
transactions
```

```python
patterns = pyfpgrowth.find_frequent_patterns(transactions,3)
rules = pyfpgrowth. generate_association_rules(patterns,0.6)
```

```python
print("PATTERNS: ", patterns)
print("RULES: ", rules)
```
