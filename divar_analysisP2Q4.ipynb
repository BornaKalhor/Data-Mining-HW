```python
df1 =df[['platform','created_at']]
df1
```

```python
gpbydate = df1.groupby(by="created_at")
transactions = []
gpbydate.filter(lambda x: (transactions.append(unique(x['platform'].to_list()))))
transactions[-10:]
```

```python
patterns = pyfpgrowth.find_frequent_patterns(transactions,3)
rules = pyfpgrowth. generate_association_rules(patterns,0.6)
```

```python
print("PATTERNS: ", patterns)
print("RULES: ", rules)
```
