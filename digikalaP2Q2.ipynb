```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import pyfpgrowth as fpgrowth
import networkx
```

```python
orders = pd.read_csv(r"/content/drive/My Drive/digikala_dataset/orders.csv")
```

```python
orders.head()
```

```python
orders.info()
```

```python
transactions = []
gpd_by_customer = orders.groupby(by=['ID_Customer'])
# transactions = [ each['ID_Item'].tolist() for each, this in gpd_by_customer ]

# for this in gpd_by_customer:
#   transactions.append(this['ID_item'].tolist())

gpd_by_customer.filter(lambda x: (transactions.append(x['ID_Item'].to_list())))
```

```python
pattern = fpgrowth.find_frequent_patterns(transactions, 3)
list(pattern)[:20]
```

```python
rules = fpgrowth.generate_association_rules(pattern, 0.75)
rules
```

```python
for each in rules: print(str(each))
```
